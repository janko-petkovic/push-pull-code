```python
import numpy as np
import torch

import matplotlib
matplotlib.rcParams['font.sans-serif'] = 'open sans'
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import matplotlib.colors as pltcolors
import matplotlib.cm as cm


import palettable.colorbrewer.diverging as paldiv

from rdn.validation import Simulation
from rdn.defaults import pardict_from_result
from rdn.fitting.models import LocalGaussModelTilde
```

# Panels b, c, and d

```python
model = LocalGaussModelTilde()

model_p_dict = pardict_from_result(
    'output/multi_fitting/Multi_LocalGaussModelTilde/NLLAdast/1_3_5_7_Spine_data_fides_1200.hdf5', 
    Chi=1, dendrite_length=1000, N_mean=5000, run_index=0)
```

```python
c_K = 'tab:green'
c_N = 'mediumvioletred'
c_P = 'cornflowerblue'
c_P = 'tab:blue'
alpha_cont = 0.2
```

Simulation

```python
simulation_time = 40
spine_number = 200
inter_spine_distance = 1

# Account for shorter dendrite used here
model_p_dict['Pi'] /= 10

# Alternative tau, as discussed
model_p_dict['tau_N'] *= 2
model_p_dict['tau_K'] *= 2


n_experiments = 20
plot_spine_idxes = torch.arange(96,105)
stim_idxes = torch.tensor([100])

simulation = Simulation(model = model,
                    model_p_dict = model_p_dict,
                    simulation_time = simulation_time,
                    spine_number = spine_number,
                    inter_spine_distance = inter_spine_distance,
                    stim_indexes = stim_idxes)

bsb, sb, rsb, kb, kbb, nb, nbb = simulation.run_return_all(n_experiments, seed=2)

# Get some potentiated and some depressed indexes
pot_idx = torch.where(rsb[3,100,:]>1)[0][:2]
dep_idx = torch.where(rsb[3,100,:]<1)[0][:2]
```

```python
def induction_profile(X, model_p_dict):
    '''given X as distance from stimulus, returns the
    two K and N induction profiles'''

    return (
        model_p_dict['Ks'] * np.exp(-X**2/model_p_dict['sigma_K']),
        model_p_dict['Ns'] * np.exp(-X**2/model_p_dict['sigma_N']),
    )
```

Plotting

```python
# Quantities to plot
X_sp = np.arange(-4,5)
X_cont = np.linspace(-4,4,100)


# Simulated values
experiment_idx = 1
K_post_sp = kb[2,96:105,experiment_idx]
N_post_sp = nb[2,96:105,experiment_idx]
K_pre_sp = kbb[2,96:105,experiment_idx]
N_pre_sp = nbb[2,96:105,experiment_idx]
K_ratio_sp = K_post_sp/K_pre_sp
N_ratio_sp = N_post_sp/N_pre_sp

P_pre_sp = bsb[2,96:105, experiment_idx]
P_post_sp = sb[2,96:105, experiment_idx]
P_ratio_sp = rsb[2,96:105,experiment_idx]

c_K = 'tab:green'
c_N = 'mediumvioletred'
c_P = 'cornflowerblue'
c_P = 'tab:blue'
alpha_cont = 0.5
```

```python
def plot_induction_profile(ax, X_cont, model_p_dict, **kwargs):

    K_cont, N_cont = induction_profile(X_cont, model_p_dict)

    ax.plot(X_cont, K_cont, c=c_K, alpha=1,
        # linestyle=(0,(10,3)),
        label='kinase K',
    )
    ax.plot(X_cont, N_cont, c=c_N, alpha=1,
        # linestyle=(0,(10,3)),
        label='phosphatase N',
    )

    ax.legend(ncols=2, frameon=False, loc=(0.03, 0.65), fontsize=8)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    ax.set_xticks([-4,-2,0,2,4])

    ax.set_ylim(0,5.9e4)
    ax.ticklabel_format(axis='y', scilimits=(0,1), useMathText=True)
    ax.yaxis.offsetText.set_position((-0.1,1))

    

def plot_catalyst_induction(axs, X_sp, C_basal, C_post, **kwargs):

    # Catalyst variation
    ax = axs[0]

    # Basal
    ax.errorbar(X_sp, C_basal, fmt='-o', c='gray',
                alpha=alpha_cont, markeredgewidth=0,
                label='basal')

    # Post
    ax.errorbar(X_sp, C_post, fmt='-o', c=kwargs.get('c'), 
                markeredgewidth=0,
                label='after induction')

    ax.legend(ncols=2, frameon=False, loc=(0.03, 0.65), fontsize=8)

    for k, v in ax.spines.items():
        if k != 'left': v.set_visible(False)

    ax.set_ylim(0,5.9e4)
    ax.set_xticks([])
    ax.ticklabel_format(axis='y', scilimits=(0,1), useMathText=True)
    ax.yaxis.offsetText.set_position((-0.1,1))


    # Post-pre ratio
    ax = axs[1]
    C_ratio = C_post / C_basal
    ax.bar(
        X_sp, C_ratio, width=0.3, color=kwargs.get('c'),
        label=kwargs.get('bar_label')
    )

    ax.legend(frameon=False, loc='center left', fontsize=8)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.set_ylim(0,13)


def plot_outcome(axs, X_sp, P_basal, P_post, K_ratio, N_ratio, **kwargs):

    # P ratio outcome

    ax = axs[0]
    ax.errorbar(X_sp, P_basal, fmt='-o', c='gray',
                alpha=alpha_cont, markeredgewidth=0,
                label='basal')

    # Post
    ax.errorbar(X_sp, P_post, fmt='-o', c=kwargs.get('c'), 
                markeredgewidth=0,
                label='after induction')

    for k, v in ax.spines.items():
        if k != 'left': v.set_visible(False)

    ax.legend(ncols=2, frameon=False, loc=(0.03, 0.65), fontsize=8)

    ax.set_ylim(0.5e4,3.5e4)
    ax.set_xticks([])
    ax.ticklabel_format(axis='y', scilimits=(0,1), useMathText=True)
    ax.yaxis.offsetText.set_position((-0.1,1))


    ax = axs[1]
    P_ratio = P_post / P_basal
    ax.axhline(y=1, linewidth=1, linestyle=(0,(5,3)), c='gray', alpha=alpha_cont)
    ax.errorbar(X_sp, P_ratio, alpha=1, c=kwargs.get('cP'), fmt='-o',
                label='size post-basal ratio')

    ax.fill_between(X_sp, 1, P_ratio, alpha=0.2)

    ax.legend(frameon=False, loc=(0.02, .62), fontsize=8)

    for k, v in ax.spines.items():
        if k != 'left': v.set_visible(False)
    ax.set_xticks([])
    ax.set_ylim(0.5,1.8)

    # Ratios of catalysts
    ax = axs[2]
    ax.bar(X_sp-0.15, K_ratio, width=0.3, color=kwargs.get('cK'), label='K post-basal ratio')
    ax.bar(X_sp+0.15, N_ratio, width=0.3, color=kwargs.get('cN'), label='N post-basal ratio')

    ax.legend(frameon=False, loc=(0.02,.3), fontsize=8)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.set_xlabel(r'Spine position relative to simulus [$\mu m$]')
    ax.set_ylim(0,13)
```

```python
fig, axs = plt.subplots(
    11,1, figsize=(4.2,11),
    height_ratios=(
        1,.3,
        1,0.5,0.3,
        1,0.5,0.3,
        1, 1, 0.5,
    ), 
    dpi=300,
)

fig.subplots_adjust(hspace=0.2, left=0.13, right=0.9, bottom=0.2)


plot_induction_profile(axs[0], X_cont, model_p_dict)

plot_catalyst_induction(
    axs[2:4],
    X_sp, K_pre_sp, K_post_sp,
    c=c_K,
    bar_label='K post-pre ratio'
)

plot_catalyst_induction(
    axs[5:7],
    X_sp, N_pre_sp, N_post_sp,
    c=c_N,
    bar_label='N post-pre ratio'
)


plot_outcome(
    axs[-3:], X_sp, P_pre_sp, P_post_sp, K_ratio_sp, N_ratio_sp,
    cP=c_P, cK=c_K, cN=c_N
)

for ax in axs[[1,4,7,]]: ax.remove()

axs[0].text(-4.1, 5.5e4, 'Catalyst induced profile', fontweight='bold')
axs[2].text(-4.1, 5.5e4, 'Synaptic kinase amount', fontweight='bold')
axs[5].text(-4.1, 5.5e4, 'Synaptic phosphatase amount', fontweight='bold')
axs[-3].text(-4.1, 3.4e4, 'Synaptic size plasticity profile', fontweight='bold')

# plt.savefig('figures-paper/single-panels/mechanism-breakdown.svg')
```

```python
cs_grad = [
    'tab:blue',
    '#0096c7',
    '#90e0ef',
    '#caf0f8',
]

def get_y_colors(y):
    norm = pltcolors.LogNorm(vmin=0.7, vmax=1.4)
    mapper = cm.ScalarMappable(norm=norm, cmap=paldiv.RdBu_4.mpl_colormap)

    return [mapper.to_rgba(yy) for yy in y]
        

tau_K = model_p_dict['tau_K']
tau_N = model_p_dict['tau_N']


fig, axs = plt.subplots(4,1, figsize=(4.1,4.5), dpi=300)
fig.subplots_adjust(hspace=0.5, left=0.13, right=0.9, bottom=0.2)

for ax, c, t in zip(axs, cs_grad, [2,10,20,30]):

    Y = rsb[t, 96:105, experiment_idx]

    colors = get_y_colors(Y)
    ax.scatter(X_sp, Y, c=colors, linewidths=1, 
               zorder=11)
    ax.plot(X_sp, Y, c='gray', zorder=10, alpha=0.2)
    
    ax.text(-4.2,2.2, 'Post-basal ratio: {} min'.format(t), va='top', ha='left', fontsize=10)
    
    ax.set_xticks([])
    ax.set_ylim(0,2)
    ax.set_yticks([])

ax.set_xticks(X_sp[::2]);

ax.set_xlabel(r'Spine position relative to stimulus $[\mu m]$', labelpad=15)


for ax in axs:
    for s in ax.spines.values(): s.set_visible(False)
    ax.axhline(y=1, linewidth=1, linestyle=(0,(5,3)), c='gray', alpha=alpha_cont)


# plt.savefig('figures-paper/single-panels/plasticity-evolution.svg')
```

```python
t_end = 20 #in minutes

t_pre = np.array([-5,0])
t_post = np.arange(2,t_end)
t_ticks = (-5,0,10,20)

line_kwargs = dict(
    linewidth=2,
)


def plot_catalyst_induction_time(ax, t_pre, t_post, spine_position):

    i = 0

    k_post = kb[2:t_end,spine_position,i]
    n_post = nb[2:t_end,spine_position,i]
    k_pre = kbb[2, spine_position,i]
    n_pre = nbb[2, spine_position,i]


    t_mid = [t_pre[-1], t_post[0]]
    k_mid = [k_pre, k_post[0]]
    n_mid = [n_pre, n_post[0]]
    
    ax.plot(t_pre, k_pre.repeat(len(t_pre)), c=c_K, **line_kwargs)
    ax.plot(t_post, k_post, c=c_K, **line_kwargs, label='Kinases')
    ax.plot(t_mid, k_mid, c=c_K, linestyle=(0,(4,1)), **line_kwargs)

    ax.plot(t_pre, n_pre.repeat(len(t_pre)), c=c_N, **line_kwargs)
    ax.plot(t_post, n_post, c=c_N, **line_kwargs, label='Phosphatases')
    ax.plot(t_mid, n_mid, c=c_N, linestyle=(0,(4,1)), **line_kwargs)

    ax.set_xticks(t_ticks)

    ax.set_ylim(0,4.5e4)
    ax.ticklabel_format(axis='y', scilimits=(0,1), useMathText=True)
    ax.yaxis.offsetText.set_position((-.1,0))

    for k, v in ax.spines.items():
        if (
            k != 'bottom' 
            and k !='left'
        ): 
            v.set_visible(False)



fig, axs = plt.subplots(2,1, figsize=(4.2,2.), sharex=True, dpi=300)
fig.subplots_adjust(hspace=0.5, bottom=0.2)


plot_catalyst_induction_time(axs[0], t_pre, t_post, 100)
plot_catalyst_induction_time(axs[1], t_pre, t_post, 99)
# plot_catalyst_induction_time(axs[2], t_pre, t_post, 98)

axs[0].legend(frameon=False, ncols=2, loc=(0.1, 1.4))
axs[0].text(-5, 5.2e4, r'Evolution at stimulation', fontweight='bold')
axs[1].text(-5, 3.5e4, r'Evolution at distance $\mathbf{1} \, \mathbf{\mu m}$', fontweight='bold')
# axs[2].text(-5, 3e4, r'$2 \, \mu m$')

axs[1].set_xlabel('Time from induction protocol [min] ')

# plt.savefig('figures-paper/single-panels/catalyst-evolution.svg')
```
